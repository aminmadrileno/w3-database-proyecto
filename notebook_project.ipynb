```python
query=('''select title, count
        from
        (select film_id, count(rental_id) as count
        from completa
        where rental_id is not null
        group by film_id) as s
        
        left join film
        on s.film_id=film.film_id
        order by count desc
        limit 3
        ''')

query2=pd.read_sql(query, cursor)
query2
```

```python
query=('''select concat(first_name,' ', last_name) as actorname, count
        from
        (select actor_id, count(rental_id) as count
        from completa
        where rental_id is not null and first_name is not null
        group by actor_id) as s
        
        left join actor
        on s.actor_id=actor.actor_id
        order by count desc
        limit 3
        ''')
query3=pd.read_sql(query, cursor)
query3
```

```python
query=('''select average
        from
        (select avg(rental_duration) as average
        from completa
        where rental_id is not null) as s
        
        ''')

query4=pd.read_sql(query, cursor)
query4
```

```python
query=('''select name, count
        from
        (select category_id, count(rental_id) as count
        from completa
        where rental_id is not null and name is not null
        group by category_id) as s
        
        left join category
        on s.category_id=category.category_id
        order by count desc
        limit 3
        ''')

query5=pd.read_sql(query, cursor)
query5
```

```python
query=('''select name, count
        from
        (select category_id, count(rental_id) as count
        from completa
        where rental_id is not null and name is not null
        group by category_id) as s
        
        left join category
        on s.category_id=category.category_id
        order by count asc
        limit 3
        ''')

query6=pd.read_sql(query, cursor)
query6
```

```python
query=('''select staff, count
        from
        (select staff_id as staff, count(rental_id) as count
        from completa
        where rental_id is not null
        group by staff_id
        ) as s
        order by count desc
        
        limit 3
        ''')

query7=pd.read_sql(query, cursor)
query7
```
